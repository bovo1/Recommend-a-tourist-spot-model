```python
#수도권
s_df_place = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/수도권/tn_visit_area_info_방문지정보_A.csv')
s_df_travel = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/수도권/tn_travel_여행_A.csv')
s_df_traveler = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/수도권/tn_traveller_master_여행객 Master_A.csv')

#동부권
e_df_place = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/동부권/tn_visit_area_info_방문지정보_B.csv')
e_df_travel = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/동부권/tn_travel_여행_B.csv')
e_df_traveler = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/동부권/tn_traveller_master_여행객 Master_B.csv')

#서부권
w_df_place = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/서부권/tn_visit_area_info_방문지정보_C.csv')
w_df_travel = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/서부권/tn_travel_여행_C.csv')
w_df_traveler = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/서부권/tn_traveller_master_여행객 Master_C.csv')

#제주도 및 섬지역
i_df_place = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/제주도/tn_visit_area_info_방문지정보_D.csv')
i_df_travel = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/제주도/tn_travel_여행_D.csv')
i_df_traveler = pd.read_csv('/content/drive/MyDrive/추천/travelDATA/DATA/Training/LabledDATA/train_data/제주도/tn_traveller_master_여행객 Master_D.csv')
```

```python
df_place = pd.concat([s_df_place, e_df_place, w_df_place, i_df_place])
df_travel = pd.concat([s_df_travel, e_df_travel, w_df_travel, i_df_travel])
df_traveler = pd.concat([s_df_traveler, e_df_traveler, w_df_traveler, i_df_traveler])
```

```python
df_place2 = df_place[~df_place['REVISIT_YN'].isnull()].copy()
```

```python
df_place2['LOTNO_ADDR'].fillna(df_place2['ROAD_NM_ADDR'], inplace=True)
```

```python
F_SGG = df_place2['LOTNO_ADDR'].str.split(' ', expand=True)[0]
S_SGG = df_place2['LOTNO_ADDR'].str.split(' ', expand=True)[1]

df_place2['SGG_CD'] = F_SGG + ' ' + S_SGG
```

```python
df2 = pd.merge(df_travel, df_traveler, on='TRAVELER_ID', how='left')
df2 = pd.merge(df_place2, df2, on='TRAVEL_ID', how='left')
```

```python
df2
```

```python
filtered_df = df2[[
    'GENDER',
    'AGE_GRP',
    'TRAVEL_COMPANIONS_NUM',
    'VISIT_AREA_NM',
    'TRAVEL_TERM',
    'TRAVEL_PURPOSE',
    'DGSTFN',
    'INCOME',
    'TRAVEL_LIKE_SIDO_1','TRAVEL_LIKE_SIDO_2',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1', 'TRAVEL_MOTIVE_2',
  ]]
```

```python
filtered_df
```

```python
filtered_df.dropna(inplace=True)
filtered_df
```

```python
filtered_df.loc[:, 'TRAVEL_PURPOSE'] = filtered_df['TRAVEL_PURPOSE'].str.split(';').str[0].astype(int)
#filtered_df.loc[:, 'TRAVEL_PURPOSE'] = (filtered_df['TRAVEL_PURPOSE'].str.split(';').str[0].astype(int) + filtered_df['TRAVEL_PURPOSE'].str.split(';').str[1].astype(int))
```

범주형 데이터를 숫자로

```python
category_data = [
    'GENDER',
    #'AGE_GRP',
    #'TRAVEL_COMPANIONS_NUM',
    'VISIT_AREA_NM',
    'TRAVEL_TERM',
    'TRAVEL_PURPOSE',
    #'DGSTFN',
    #'INCOME',
    'TRAVEL_LIKE_SIDO_1','TRAVEL_LIKE_SIDO_2',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1', 'TRAVEL_MOTIVE_2',
]
```

```python
filtered_df[category_data[2:]] = filtered_df[category_data[2:]].astype(int)
```

```python
filtered_df
```

##훈련

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(filtered_df, test_size=0.1, random_state=42)

print(train_data.shape)
print(test_data.shape)
```

```python
!pip install -q catboost
```

```python
from catboost import CatBoostRegressor, Pool

train_pool = Pool(train_data.drop(['DGSTFN'], axis=1),
                  label = train_data['DGSTFN'],
                  cat_features = category_data)

test_pool = Pool(test_data.drop(['DGSTFN'], 1),
                 label = test_data['DGSTFN'],
                 cat_features = category_data)
```

```python
model = CatBoostRegressor(
    loss_function='RMSE',
    eval_metric='MAE',
    task_type='GPU',
    depth=12,
    learning_rate=0.001,
    n_estimators=10000)

model.fit(
    train_pool,
    eval_set=test_pool,
    verbose=1000,
    plot=True)
```

```python
model.get_best_score()
```
